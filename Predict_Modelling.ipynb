```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = pd.read_excel('NBA_Stats_71_Years.xlsx')
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

### Replacing all null values with their mean

```python
df['3-Pt FG Attempts'] = df['3-Pt FG Attempts'].fillna(df['3-Pt FG Attempts'].mean())
df['3-Pt FG %'] = df['3-Pt FG %'].fillna(df['3-Pt FG %'].mean())
df['3-Pt FG Made'] = df['3-Pt FG Made'].fillna(df['3-Pt FG Made'].mean())
df['Offensive Rebounds'] = df['Offensive Rebounds'].fillna(df['Offensive Rebounds'].mean())
df['Defensive Rebounds'] = df['Defensive Rebounds'].fillna(df['Defensive Rebounds'].mean())
df['Steals'] = df['Steals'].fillna(df['Steals'].mean())
df['Blocks'] = df['Blocks'].fillna(df['Blocks'].mean())
df['Turnovers'] = df['Turnovers'].fillna(df['Turnovers'].mean())
df['AST/TOV'] = df['AST/TOV'].fillna(df['AST/TOV'].mean())
df['STL/TOV'] = df['STL/TOV'].fillna(df['STL/TOV'].mean())
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df['Season Type'] = df['Season Type'].astype('category')
df['Season Type'] = df['Season Type'].cat.codes
df
```

```python
#Training Features
X = df.drop(columns = ['Rank', 'Season Start Year', 'Year', 'Player', 'Team', 'Team ID', 'Player ID'])
X
```

```python
#Label
y = df['Rank']
```

```python
from sklearn.model_selection import train_test_split
```

```python
#Splitting our dataset using 30% of our data as test zize
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 21)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#Fitting our training data
lr = LinearRegression().fit(X_train, y_train)
```

```python
# Y intercept
c = lr.intercept_
```

```python
c
```

```python
# Coeficient
m = lr.coef_
m
```

```python
#Prediction using our training data
y_pred_train = lr.predict(X_train)
y_pred_train
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_train, y_pred_train)
plt.xlabel('Actual Rank')
plt.ylabel('Predicted Rank')
plt.show()
```

```python
from sklearn.metrics import r2_score
```

```python
#Checking R2 Score
r2_score(y_train, y_pred_train)
```

```python
#Predict using our test data
y_pred_test = lr.predict(X_test)
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_test, y_pred_test)
plt.xlabel('Actual Rank')
plt.ylabel('Predicted Rank')
plt.show()
```

```python
r2_score(y_test, y_pred_test)
```
