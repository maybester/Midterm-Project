## Exploratory Data Analysis

```python
# get the player with the highest average points per game in a specific season over history
def get_ranked_players(seasons, players, points, games):
    player_averages = {}

    # Calculate average points for each player in each season
    for season, player, point, game in zip(seasons, players, points, games):
        if player in player_averages:
            player_averages[player].append((season, point / game))
        else:
            player_averages[player] = [(season, point / game)]

    ranked_players = []
    for player, season_averages in player_averages.items():
        season_averages.sort(key=lambda x: x[1], reverse=True)
        for season, average_score in season_averages:
            ranked_players.append((player, season, average_score))

    ranked_players.sort(key=lambda x: x[2], reverse=True)

    return ranked_players

```

```python
# list of players with highest scored in history and the season they performed
seasons = df_regular_seasons['Season Start Year']
players = df_regular_seasons['Player']
points = df_regular_seasons['Points Scored']
games = df_regular_seasons['Games Played']
ranked_players_rs = get_ranked_players(seasons, players, points, games)
ranked_players_rs[:10]
```

```python
# output in csv file
df_highiest_score = pd.DataFrame(ranked_players_rs, columns=['Player', 'Season', 'Average Score'])
# df_highiest_score.to_csv('player_score.csv', index=False)
```

```python
# player with the most assists in a given season
def get_ranked_assist(seasons, players, assists):
    player_averages = {}

    # Calculate assists for each player in each season
    for season, player, assist in zip(seasons, players, assists):
        if player in player_averages:
            player_averages[player].append((season, assist))
        else:
            player_averages[player] = [(season, assist)]

    ranked_players = []
    for player, season_averages in player_averages.items():
        for season, average_assist in season_averages:
            ranked_players.append((player, season, average_assist))

    ranked_players.sort(key=lambda x: x[2], reverse=True)

    return ranked_players

```

```python
# list of players with highest assists score in history and the season they performed
seasons = df_regular_seasons['Season Start Year']
players = df_regular_seasons['Player']
assists = df_regular_seasons['Assists']
ranked_assists_rs = get_ranked_assist(seasons, players, assists)
ranked_assists_rs[:10]
```

```python
# output csv file
df_assists_rs = pd.DataFrame(ranked_assists_rs, columns=['Player', 'Season', 'Assists Score'])
# df_assists_rs.to_csv('player_assists.csv', index=False)
```

```python
#rank the best rebounder in a specific league (team)
def get_ranked_rebounder(teams, players, rebounds):
    team_rebounds = {}

    # Calculate highest rebound score for each team
    for team, player, rebound in zip(teams, players, rebounds):
        if team in team_rebounds:
            if rebound > team_rebounds[team][1]:
                team_rebounds[team] = (player, rebound)
        else:
            team_rebounds[team] = (player, rebound)

    sorted_teams = sorted(team_rebounds.items(), key=lambda x: x[1][1], reverse=True)

    return sorted_teams
```

```python
# list of NBA teams with the players who scored highest rebounds in history
teams = df_regular_seasons['Team ID']
players = df_regular_seasons['Player']
rebounds = df_regular_seasons['Rebounds']
ranked_rebounds_rs = get_ranked_rebounder(teams, players, rebounds)
ranked_rebounds_rs[:10]
```

```python
df_rebounds_rs = pd.DataFrame(ranked_rebounds_rs, columns=['Team ID', 'Player'])
df_rebounds_rs[['Player Name', 'Rebounds score']] = pd.DataFrame(df_rebounds_rs['Player'].tolist())
df_rebounds_rs.drop('Player', axis=1, inplace=True)
df_rebounds_rs = df_rebounds_rs[['Player Name', 'Rebounds score', 'Team ID']]
df_rebounds_rs
```

```python
def get_players_with_fg_counts(players, fg_rates):

# Create a dictionary to store the counts for each player
    player_counts = {}

    # Iterate over the players and FG rates
    for player, rate in zip(players, fg_rates):
        if player not in player_counts:
            player_counts[player] = {'1pct': 0, '0to0_5': 0, '0_5to1': 0}

        if rate == 1.0:
            player_counts[player]['1pct'] += 1
        elif rate > 0.0 and rate < 0.5:
            player_counts[player]['0to0_5'] += 1
        elif rate >= 0.5 and rate < 1.0:
            player_counts[player]['0_5to1'] += 1

    sorted_players = sorted(player_counts.items(), key=lambda x: x[1]['1pct'], reverse=True)

    return sorted_players

```

```python
# list of NBA teams with players performed most efficient shoot (based on FG%) in history
teams = df_regular_seasons['Team ID']
players = df_regular_seasons['Player']
fg_rates = df_regular_seasons['FG %']
ranked_FGS_rs = get_players_with_fg_counts(players, fg_rates)
ranked_FGS_rs
```

```python
ranked_FGS_df = pd.DataFrame(ranked_FGS_rs, columns=['Player', 'Values'])

# Expand the 'Values' dictionary into separate columns
ranked_FGS_df = pd.concat([ranked_FGS_df['Player'], ranked_FGS_df['Values'].apply(pd.Series)], axis=1)
ranked_FGS_df
```

```python
def find_team_with_highest_avg_points(teams, players, points, games, seasons):
    team_avg_points = {}  # Dictionary to store the average points per team for each season
    
    # Iterate over the players' data
    for team, player, pts, gms, season in zip(teams, players, points, games, seasons):
        key = (team, season)
        if key not in team_avg_points:
            team_avg_points[key] = {'total_pts': 0, 'total_gms': 0, 'num_players': 0}

        team_avg_points[key]['total_pts'] += pts
        team_avg_points[key]['total_gms'] += gms
        team_avg_points[key]['num_players'] += 1

    highest_avg_teams = []

    # Calculate average points per team for each season
    for key, data in team_avg_points.items():
        team, season = key
        total_pts = data['total_pts']
        total_gms = data['total_gms']
        num_players = data['num_players']
        avg_points = total_pts / total_gms / num_players

        highest_avg_teams.append((season, team, avg_points))

    # Sort the teams based on average points per game in descending order
    highest_avg_teams = sorted(highest_avg_teams, key=lambda x: x[2], reverse=True)

    return highest_avg_teams

```

```python
# team has the highest average points per game in a specific season
teams = df_regular_seasons['Team ID']
players = df_regular_seasons['Player']
points = df_regular_seasons['Points Scored']
games = df_regular_seasons['Games Played']
season = df_regular_seasons['Season Start Year']

result = find_team_with_highest_avg_points(teams, players, points, games, seasons)
result
```
